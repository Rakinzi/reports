```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "China": 99,
    "Singapore": 55,
    "Zimbabwe": 38,
    "United States": 25,
    "South Africa": 7,
}

df = pd.Series(data)

bar_color = "#156082"

# Format function (1 decimal for k)
def format_k(value):
    if value >= 1000:
        return f"{value/1000:.1f}k".replace(".", ",")
    return str(value)

plt.figure(figsize=(8, 6))
bars = plt.bar(df.index, df.values, color=bar_color)

# Add formatted values above bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + max(df.values) * 0.02,  # dynamic spacing
        format_k(height),
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

# Remove axes and borders
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(False)
ax.get_yaxis().set_visible(False)

plt.tick_params(left=False, bottom=False)

plt.xticks(rotation=0, fontsize=10)

plt.tight_layout()

plt.savefig("blue_chart.png", dpi=300, bbox_inches='tight', transparent=True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data (only two sources)
data = {
    "Direct": 188,
    "Organic Search": 30
}

# Convert to pandas Series
df = pd.Series(data)

# Colors swapped: Direct = Orange, Organic Search = Blue
colors = ["#e97132", "#156082"]

# Create figure
plt.figure(figsize=(6, 6))

# Plot pie chart
wedges, _ = plt.pie(
    df.values,
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1}
)

# Title
plt.title("Source of Traffic", fontsize=14, pad=5)

# Legend below chart
plt.legend(
    df.index,
    loc="upper center",
    bbox_to_anchor=(0.5, -0.05),
    ncol=2,
    frameon=False,
    fontsize=10
)

# Adjust spacing
plt.subplots_adjust(top=0.9, bottom=0.15)

# Save image
plt.savefig("traffic_pie_chart_compact.png", dpi=300, bbox_inches='tight', transparent=True)

# Show chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "New Users": 225,
    "Returning Users": 14
}

df = pd.Series(data)

colors = ["#156082", "#e97132"]  # Blue, Orange

# Proper formatter (always 1 decimal for k)
def format_k(value):
    if value >= 1000:
        return f"{value/1000:.1f}k".replace(".", ",")
    return str(value)

plt.figure(figsize=(6, 6))

wedges, texts, autotexts = plt.pie(
    df.values,
    labels=None,
    autopct=lambda p: format_k(int(round(p * sum(df.values) / 100))),
    startangle=90,
    colors=colors,
    textprops={'fontsize': 14, 'color': 'white', 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 0}
)

plt.legend(
    df.index,
    loc="upper center",
    bbox_to_anchor=(0.5, -0.05),
    ncol=2,
    frameon=False,
    fontsize=10
)

plt.subplots_adjust(top=0.9, bottom=0.15)

plt.savefig("user_pie_chart.png", dpi=300, bbox_inches='tight', transparent=True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "One": 43,
    "Two": 68,
    "Three": 57,
    "Four": 63
}

df = pd.Series(data)

# Format function (1 decimal for k)
def format_k(value):
    if value >= 1000:
        return f"{value/1000:.1f}k".replace(".", ",")
    return str(value)

plt.figure(figsize=(7, 5))

color = "#156082"

# Plot line with circle markers
plt.plot(df.index, df.values, color=color, linewidth=2.5, marker='o', markersize=8)

# Add formatted values above each point
for i, value in enumerate(df.values):
    plt.text(
        i,
        value + max(df.values) * 0.03,  # dynamic spacing above point
        format_k(value),
        ha='center',
        va='bottom',
        fontsize=22,
        fontweight='bold'
    )

# Remove y-axis and all spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.get_yaxis().set_visible(False)

plt.tick_params(left=False, bottom=False)
plt.xticks(fontsize=22)

plt.tight_layout()

plt.savefig("line_chart.png", dpi=300, bbox_inches='tight', transparent=True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Home": 107,
    "About": 42,
    "Living With Cancer": 17,
    "Activities": 11
}

df = pd.Series(data)

bar_color = "#156082"

# Format function (1 decimal for k)
def format_k(value):
    if value >= 1000:
        return f"{value/1000:.1f}k".replace(".", ",")
    return str(value)

# Create figure
plt.figure(figsize=(12, 6))
bars = plt.bar(df.index, df.values, color=bar_color, width=0.6)

# Add formatted values on top
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 20,  # slightly higher offset for better spacing
        format_k(height),
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

# Remove axes, borders, and ticks
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(False)
ax.get_yaxis().set_visible(False)
plt.tick_params(left=False, bottom=False)

# X labels
plt.xticks(fontsize=14, rotation=20)  # slight rotation for readability

plt.tight_layout()

plt.savefig("page_views_chart.png", dpi=300, bbox_inches='tight', transparent=True)

plt.show()
```
